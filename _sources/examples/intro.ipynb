# Examples Outline
The following examples show working code that demonstrate many of Aviary's features. They typically demonstrate a full working aircraft design optimization to showcase a "complete" Aviary problem.

The examples slowly ramp up in complexity and build upon knowledge from previous examples and the documentation. Each example lists relevant pages that are recommended for users to read for more information about the key features introduced in that example.
